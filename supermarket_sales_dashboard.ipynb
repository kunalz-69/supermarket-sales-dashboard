```python
# ðŸ“¦ Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ“¥ Load Dataset
df = pd.read_csv("supermarket_sales_for_kunal.csv")

# ðŸ§¹ Data Overview
print("Dataset Shape:", df.shape)
df.info()
print("\nMissing values:\n", df.isnull().sum())
display(df.describe())
display(df.head())

# ðŸ§  Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# ðŸ“Š 1. Sales by Gender
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title("Sales Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Sales")
plt.show()

# ðŸ“Š 2. Payment Method Usage
plt.figure(figsize=(6,4))
sns.countplot(x='Payment', data=df, palette='pastel')
plt.title("Payment Method Usage")
plt.xlabel("Payment Method")
plt.ylabel("Frequency")
plt.show()

# ðŸ“Š 3. Average Rating per Product Line
plt.figure(figsize=(10,6))
sns.barplot(x='Product line', y='Rating', data=df, palette='viridis')
plt.title("Average Rating by Product Line")
plt.xticks(rotation=45)
plt.ylabel("Average Rating")
plt.show()

# ðŸ“Š 4. Total Sales by Branch
plt.figure(figsize=(6,4))
sns.barplot(x='Branch', y='Total', data=df, estimator=sum, palette='cool')
plt.title("Total Sales by Branch")
plt.ylabel("Total Revenue")
plt.show()

# ðŸ“ˆ 5. Daily Sales Trend
daily_sales = df.groupby('Date').sum(numeric_only=True)['Total']
plt.figure(figsize=(12,5))
daily_sales.plot(marker='o')
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```
